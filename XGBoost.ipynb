```python
import pandas as pd
import numpy as np
import os

import xgboost as xgb
from xgboost import XGBClassifier

from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, classification_report

import pickle
import joblib
```

```python
train = pd.read_csv("train_mwmote.csv")
```

```python
y = train['target']
x = train.drop('target', axis = 1)
```

# Trials

## 1st trial

```python
xgb_clf = XGBClassifier()

xgb_param_grid = {'learning_rate' : [0.01],
                  'max_depth': [4],
                  'min_child_weight' : [2.5],
                  'gamma' : [2.5],
                  'subsample': [0.7], 
                  'colsample_bytree': [0.7],
                  'reg_lambda': [2],
                  'objective': ['binary:logistic']}

xgb_grid = GridSearchCV(estimator=xgb_clf,
                       param_grid=xgb_param_grid,
                       scoring='f1',
                       n_jobs=-1,
                       cv=5,
                       refit=True, 
                       return_train_score=True)

xgb_grid.fit(x, y)
```

```python
xgb_grid.best_estimator_
```

```python
xgb_grid.best_score_
```

## 2nd trial

```python
xgb_clf = XGBClassifier()

xgb_param_grid = {'learning_rate' : [0.01],
                  'n_estimators' : [400],
                  'max_depth': [4],
                  'min_child_weight' : [2.5],
                  'gamma' : [2.5],
                  'subsample': [0.5], 
                  'colsample_bytree': [0.5],
                  'reg_lambda': [3],
                  'objective': ['binary:logistic']}

xgb_grid = GridSearchCV(estimator=xgb_clf,
                       param_grid=xgb_param_grid,
                       scoring='f1',
                       n_jobs=-1,
                       cv=5,
                       refit=True, 
                       return_train_score=True)

xgb_grid.fit(x, y)
```

```python
xgb_grid.best_score_
```

```python
xgb_grid.best_estimator_
```

```python
model = XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=0.5, gamma=2.5,
              learning_rate=0.01, max_delta_step=0, max_depth=4,
              min_child_weight=2.5, missing=None, n_estimators=400, n_jobs=1,
              nthread=None, objective='binary:logistic', random_state=0,
              reg_alpha=0, reg_lambda=3, scale_pos_weight=1, seed=None,
              silent=None, subsample=0.5, verbosity=1)
```

```python
model.fit(x, y)
```

```python
test = pd.read_csv("test_pca.csv")
```

```python
test.shape
```

```python
prediction = model.predict(test)
```

```python
submission = pd.DataFrame(columns = ['id', 'target'])
submission['id'] = range(1,2001)
```
